## Dataset Overview

An automotive corporation from India named XYZ wants to enter the country's used-car industry by establishing a local business to compete with its rivals. Since these may differ greatly from the market for new automobiles, they want to understand the factors influencing car pricing and sales. In essence, the company seeks to understand: What factors are important in determining the price of a used car? How well such factors reflect a car's price The consulting company has compiled a sizable dataset of various makes and models of used automobiles available on the market based on several market surveys. The dataset includes the different variables that will be helpful during the analysis phase, Sales ID, name, year, selling price, kilometers traveled, state or province, region, fuel type, owner, mileage, engine, maximum power, and number of seats sold. The data was obtained from https://www.kaggle.com.

This data has no missing values and duplicates so I went straight into the analysis. 
This analysis is in two parts. The first part is about exploration of the dataset and the second part is about visualization.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
cars_df = pd.read_csv('UserCarData.csv')
cars_df.head(3)
```

## Relationship between Year, Selling price and Sold

This plot describes how the selling price of used cars have appreciated over the years. Clearly, there has been a consistent increase in the selling price started from the year 2002. 2019 saw the highest selling price for used cars but 2020 saw a decline in the selling price of used cars. This might be caused by factors that might not be present in this data. The global pandemic (Covid-19) affected most businesses and that can be a major factor behind the declined in the selling price of used cars.

```python
def multi_plot(var1):
    plt.figure(figsize =(12,8))
    me = plt.gca()
    me.ticklabel_format(useOffset=False, style='plain')
    me.set_title(f'{var1} sold cars vs Selling Price')             
    me.set_xlabel('year');
    plt.xticks(rotation=60)
    sb.barplot(data = cars_df,x = var1,y = 'selling_price',hue ='sold',palette = 'plasma');
    me.set_ylabel('Selling Price ($)')

# plot a graph to display how year and sold affects selling price
multi_plot('year')
```

## Relationship between Name, Selling price and Sold

This describes the brand name that has the highest selling price. Brand names like Maruti, Hyundai and Mahindra turned out to be the most sold cars across board but further investigation shows that it might be because of their low selling price. Brand name like Lexus which has the highest selling price appears to be one of the less selling used cars on the market. This means cars with low selling price are prefered to cars with high selling price.

```python
# plot to display the relationship showing which brand names sold 
plt.figure(figsize = [20,12])
sb.countplot(data = cars_df,x = 'name',hue ='sold',order =cars_df['name'].value_counts().index)
me = plt.gca()
me.set_ylabel('Number of Cars')
me.set_xlabel('Brand name')
me.set_title('Brand name vs Sold')
# rotate x-axis ticks vertically
plt.xticks(rotation=60);
```

```python
# plot that shows the relationship between brand name, sold and selling price
plt.figure(figsize =(12,8))
me = plt.gca()
me.ticklabel_format(useOffset=False, style='plain')
me.set_title('Sold Car Name vs Selling Price')
plt.xticks(rotation = 90)
sb.barplot(data = cars_df,x = 'name',y = 'selling_price',hue ='sold',palette = 'plasma');
me.set_ylabel('Selling Price ($)')             
me.set_xlabel('Car Names');
```

## Relationship between Region and Selling price

Here, we look at the region with the highest selling price. We first investigated whether there are differences in the prices of cars across the four regions and clearly there are differences in the selling price for cars in different regions. For some cars the differences in selling price across the four region is neglible. Cars like jeep have significant difference in their selling price across the regions. At the end, the South region turned out to be the region with the highest selling price for used cars.



```python
plt.figure(figsize =(20,12))
me = plt.gca()
me.ticklabel_format(useOffset=False, style='plain')
me.set_title(' Region and Brand name vs Selling Price')
sb.barplot(data = cars_df,x = 'name',y = 'selling_price',hue ='Region',palette = 'plasma');
me.set_ylabel('Selling Price ($)')             
me.set_xlabel('Brand name');
plt.xticks(rotation=90);
```

```python
# plot graph
plt.figure(figsize =(20,12))
me = plt.gca()
me.ticklabel_format(useOffset=False, style='plain')
me.set_title(' Region and Brand name vs Selling Price')
sb.barplot(data = cars_df,x = 'Region',y = 'selling_price',hue ='sold',palette = 'plasma');
me.set_ylabel('Selling Price ($)')             
me.set_xlabel('Region');
plt.xticks(rotation=90);
```
